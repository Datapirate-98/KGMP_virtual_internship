```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Handeling the Null values
```

```python
data = pd.read_excel("transaction.xlsx")
data
```

```python
#Analysing the null values

data.count()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis = 0, subset=["online_order"])
data.isnull().sum()
```

```python
#removing the offline orders and the cacelled orders

data = data[data["online_order"] != 0]
data = data[data["order_status"] != "Cancelled"]
data["order_status"].describe()
data["online_order"].describe()
```

```python
#removing the NaN values

data.isnull().sum()
```

```python
data.corr()
```

```python
#filling all the NaN values with solex as it has more number freq
data["brand"].describe()
data["brand"] = data["brand"].fillna(method = "ffill")
```

```python
data["product_line"].describe()
data["product_line"] = data["product_line"].fillna(method = "bfill")
```

```python
data["product_class"].describe()
data["product_class"] = data["product_class"].fillna(method = "bfill")
```

```python
data["product_size"].describe()
data["product_size"] = data["product_size"].fillna(method = "ffill")
```

```python
data["standard_cost"].describe()
```

```python
#replacing the NaN valuse with Mean value

data["standard_cost"] = data["standard_cost"].fillna(data["standard_cost"].mean())
data["product_first_sold_date"] = data["product_first_sold_date"].fillna(data["product_first_sold_date"].mean())
data["Profit"] = data["Profit"].fillna(data["Profit"].mean())
```

```python
data.to_csv("Cleaned_Transaction.csv", index=False)
```
