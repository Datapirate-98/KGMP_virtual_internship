```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
data = pd.read_excel("CostumerDemographic.xlsx")
data.head(5)
```

```python
#Analysing the null values

data.isnull().sum()
```

```python
data["last_name"] = data["last_name"].fillna(method = "ffill")
data["DOB"] = data["DOB"].fillna(method = "ffill")
data_new = data.drop(['default', 'job_title'], axis=1)
```

```python
data_new.isnull().sum()
```

```python
data_new.info()
```

```python
#Handleing the NULL values of job_industry_category  and  tenure

data_new.describe()
```

```python
data_new["tenure"] = data_new["tenure"].fillna(data_new["tenure"].mean()) #filling the NaN values with 6 as per the above data_new description
data_new["job_industry_category"] = data_new["job_industry_category"].fillna(method = "ffill")
data_new.isnull().sum()
```

```python
data_new.head()
```

```python
#Now correcting the 3 gender columns Female, Femal to F and Male to M

data_new["gender"] = data_new["gender"].replace(to_replace = "Female", value = "F")
data_new["gender"] = data_new["gender"].replace("Femal", "F")
data_new["gender"] = data_new["gender"].replace("Male", "M")
data_new.head()
```
