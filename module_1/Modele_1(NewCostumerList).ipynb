```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
data = pd.read_excel("NewCostumerList.xlsx")
data.head(5)
```

```python
data.drop(["Unnamed: 16", "Unnamed: 17", "Unnamed: 18", "Unnamed: 19"], axis = 1, inplace=True)
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data["last_name"] = data["last_name"].fillna(method = "ffill")
data["DOB"] = data["DOB"].fillna(method = "ffill")
data.drop(["job_title"], axis=1, inplace=True)
```

```python
data["job_industry_category"] = data["job_industry_category"].fillna(method = "ffill")
```

```python
data.to_csv("Cleaned_NewCostumerList.csv", index=False)
```
