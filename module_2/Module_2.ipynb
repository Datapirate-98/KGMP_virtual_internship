```python
import numpy as np
import pandas as pd
import sweetviz
from pandas_profiling import ProfileReport
```

```python
data_file = pd.ExcelFile('Cleaned_Module1_Data.xlsx')
data_file.sheet_names
```

```python
data = pd.read_excel('Cleaned_Module1_Data.xlsx', sheet_name="Cleaned_CostumerDemographic")
data1 = pd.read_excel('Cleaned_Module1_Data.xlsx', sheet_name="Cleaned_NewCostumerList")
```

```python
data.head()
```

```python
data1.head()
```

# analysing the constumer Demographic data

```python

profile = ProfileReport(data, title='constumer Demographic Report', explorative=True) #analysing using pandas_profiler
```

```python
profile.to_widgets()
```

```python
profile.to_file("CostumerDemographic_Report.html")
```

# analysing the new costumer list data(data1)

```python
profile1 = ProfileReport(data1, title='New Costumer list Report', explorative=True) #analysing using pandas_profiler
```

```python
profile1.to_widgets()
```

```python
profile1.to_file("NewCostumerList_Report.html")
```

```python
data_new = data.drop(["DOB", "deceased_indicator", "customer_id"], axis = 1)
data1_new = data1.drop(["DOB", "deceased_indicator"], axis = 1)
```

```python
compared_report = sweetviz.compare([data_new, "constumer Demographic Report"], [data1_new, "New Costumer list Report"])
```

```python
compared_report.show_html("Compared_Report.html")
```

```python
my_reports = sweetviz.compare_intra(data_new, data_new["gender"] == "F", ["Male", "Female"])
```

```python
my_reports.show_html("my_Reports.html")
```
